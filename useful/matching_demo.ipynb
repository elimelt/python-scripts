```python
import numpy as np
from matching import Player
from matching.games import StableRoommates


def availability_in_common(a1, a2):
  assert(len(a1) == len(a2) and len(a1[0]) == len(a2[0]))
  n, m = len(a1), len(a1[0])
  count = 0
  for i in range(n):
    for j in range(m):
      if a1[i][j] == 1 and  a2[i][j] == 1:
        count += 1
  return count

def make_rankings(data):
  n = len(data)
  scores = np.matrix([
    [
        availability_in_common(data[i]["availability"], data[j]["availability"])
        for j in range(n)
    ] for i in range(n)
  ])

  idx = np.argsort((-1) * scores, axis=1)
  return idx.tolist()

def make_players(data):
  user_ids = [u["user_id"] for u in data]
  rankings = make_rankings(data)

  players = [
      Player(uid)
      for uid in user_ids
  ]

  for i, p in enumerate(players):
    p.set_prefs([players[idx] for idx in rankings[i] if idx != i])

  return players

def pair(data):
  players = make_players(data)
  game = StableRoommates(players)
  return game.solve()







```
